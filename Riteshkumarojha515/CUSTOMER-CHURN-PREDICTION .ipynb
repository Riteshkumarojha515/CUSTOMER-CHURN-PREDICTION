```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv(r"Churn_Modelling_with_nulls.csv")
```

```python
a.shape
```

```python
a.head()
```

```python
b=a.isnull().sum()/a.shape[0]*100
```

```python
b
```

```python
mask=b>20
mask
```

```python
c=b[mask].index
c
```

```python
a=a.drop( columns=c)## removing columns
a.head()
```

```python
for i in a.select_dtypes(include=['int','float']):
    a[i]=a[i].replace(np.nan,a[i].median())
```

```python
for i in a.select_dtypes(include=['object']):
    a[i]=a[i].replace(np.nan,a[i].mode()[0])
```

```python
a.isnull().sum().sum()
```

## encoding

```python
a.columns
```

```python
a=a.drop(columns='Surname')
```

```python
a.head()
```

```python
a=a.drop(columns='RowNumber')
```

```python
a.shape
```

```python
a=a.drop(columns='CustomerId')
```

```python
a.shape
```

```python
# one hot encoding
```

```python
a_encoded=pd.get_dummies(a,dtype='int',drop_first=True)
```

```python
a_encoded
```

```python
# train test data split
```

```python
X=a_encoded.drop(columns=['Exited'])
y=a_encoded['Exited']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=12)
```

```python
X_test.shape
```

```python
## feature scalling
```

```python
ap1=ss.fit_transform(X_train)
```

```python
ap1
```

```python
len(ap1)
```

```python
ap2=ss.transform(X_test)
```

```python
ap2
```

```python
X_Train_scale=pd.DataFrame(ap1,columns=X_train.columns)
```

```python
X_Test_scale=pd.DataFrame(ap2,columns=X_test.columns)
```

```python
X_Train_scale.shape
```

```python
X_Test_scale.shape
```

```python
import tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow  import keras
```

```python
model=Sequential()
```

```python
model.add(Dense(3, activation='sigmoid', input_shape=(11,)))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy' , optimizer='adam',metrics=['accuracy'])          # adam (calculate the weight)
```

```python
model.fit(X_Train_scale,y_train, epochs=20)        ## model is now trained
```

```python
## model predict
```

```python
y1=model.predict(X_Test_scale)
```
